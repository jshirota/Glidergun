```
Raster Tick(Raster r) {
    var count = r.Focal(Statistics.Sum, 3, 3) - r;
    return (r == 1 & count == 2 | count == 3).Con(1, 0);
}

Enumerable.Range(0, 3).Aggregate(new [] { gosper.Randomize() > 0.9 },
    (r, i) => r.Concat(new [] { Tick(r[i]) }).ToArray())
```
